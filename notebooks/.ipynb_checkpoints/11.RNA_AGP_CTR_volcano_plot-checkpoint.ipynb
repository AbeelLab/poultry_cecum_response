```python
import pandas as pd
import matplotlib
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np
import os

HERE = os. getcwd()
DATA_DIRECTORY = os.path.abspath(os.path.join(HERE, os.pardir, 'data'))
RESULT_DIRECTORY = os.path.abspath(os.path.join(HERE, os.pardir, 'results'))
OUTPUT_PATH = os.path.join(RESULT_DIRECTORY, 'DeSeq2','meta_species')
```

```python
cecum_df = pd.read_excel(os.path.join(OUTPUT_PATH, "Cecum_Digesta_35_DC_CTR_shrunk_stat_results.xlsx"),index_col=0)
colon_df = pd.read_excel(os.path.join(OUTPUT_PATH, "Colon_Digesta_35_DC_CTR_shrunk_stat_results.xlsx"),index_col=0)

```

```python
%pip install adjustText
from adjustText import adjust_text

matplotlib.rcParams.update({'font.size': 7.4})


# highlight down- or up- regulated genes
filtered_cecum_df = cecum_df[((cecum_df['log2FoldChange'] < -2) | (cecum_df['log2FoldChange'] > 2)) & (cecum_df['padj'] < 0.01)]

# padj_cecum_df = cecum_df[(cecum_df['log2FoldChange'] >= -2) & (cecum_df['log2FoldChange'] <= 2) & (cecum_df['padj'] < 0.01)]

# log2fc_cecum_df = cecum_df[((cecum_df['log2FoldChange'] < -2) | (cecum_df['log2FoldChange'] > 2)) & (cecum_df['padj'] >= 0.01)]



plt.scatter(x=filtered_cecum_df['log2FoldChange'],y=filtered_cecum_df['padj'].apply(lambda x:-np.log10(x)),s=4,label="padj & log2FC",color="Red")
# plt.scatter(x=padj_cecum_df['log2FoldChange'],y=padj_cecum_df['padj'].apply(lambda x:-np.log10(x)),s=4,label="padj",color="Blue")
# plt.scatter(x=log2fc_cecum_df['log2FoldChange'],y=log2fc_cecum_df['padj'].apply(lambda x:-np.log10(x)),s=4,label="log2FC",color="lime")
plt.scatter(x=cecum_df['log2FoldChange'],y=cecum_df['padj'].apply(lambda x:-np.log10(x)),s=1,label="Not significant",color="grey")
# plt.scatter(x=up['log2FoldChange'],y=up['padj'].apply(lambda x:-np.log10(x)),s=3,label="log10padj&log2FoldChange",color="Red")



texts=[]
for i,r in filtered_cecum_df.iterrows():
    if r['log2FoldChange'] > 2:
        texts.append(plt.text(x=r['log2FoldChange'],y=-np.log10(r['padj']),s=i,color='Orange'))
    elif r['log2FoldChange'] < -2: 
        texts.append(plt.text(x=r['log2FoldChange'],y=-np.log10(r['padj']),s=i,color='Blue'))

# for i,r in filtered_cecum_df.iterrows():
#     texts.append(plt.text(x=r['log2FoldChange'],y=-np.log10(r['padj']),s=i))
# for i,r in padj_cecum_df.iterrows():
#     texts.append(plt.text(x=r['log2FoldChange'],y=-np.log10(r['padj']),s=i))
# for i,r in log2fc_cecum_df.iterrows():
#     texts.append(plt.text(x=r['log2FoldChange'],y=-np.log10(r['padj']),s=i))
    
    
adjust_text(texts,arrowprops=dict(arrowstyle="-", color='black', lw=0.2))



plt.xlabel("log2FC")
plt.ylabel("log10padj")
plt.axvline(-2,color="grey",linestyle="--", lw=0.3)
plt.axvline(2,color="grey",linestyle="--", lw=0.3)
plt.axhline(2,color="grey",linestyle="--", lw=0.3)
# plt.legend(loc='upper right')
plt.savefig(os.path.abspath(os.path.join(HERE, os.pardir, 'results', 'DeSeq2','meta_species', 'Cecum_Digesta_35_DC_CTR_volcano.png')))
```

```python
%pip install adjustText
from adjustText import adjust_text

matplotlib.rcParams.update({'font.size': 7.4})


# highlight down- or up- regulated genes
filtered_colon_df = colon_df[((colon_df['log2FoldChange'] < -2) | (colon_df['log2FoldChange'] > 2)) & (colon_df['padj'] < 0.01)]

# padj_colon_df = colon_df[(colon_df['log2FoldChange'] >= -2) & (colon_df['log2FoldChange'] <= 2) & (colon_df['padj'] < 0.01)]

# log2fc_colon_df = colon_df[((colon_df['log2FoldChange'] < -2) | (colon_df['log2FoldChange'] > 2)) & (colon_df['padj'] >= 0.01)]



plt.scatter(x=filtered_colon_df['log2FoldChange'],y=filtered_colon_df['padj'].apply(lambda x:-np.log10(x)),s=4,label="padj & log2FC",color="Red")
# plt.scatter(x=padj_colon_df['log2FoldChange'],y=padj_colon_df['padj'].apply(lambda x:-np.log10(x)),s=4,label="padj",color="Blue")
# plt.scatter(x=log2fc_colon_df['log2FoldChange'],y=log2fc_colon_df['padj'].apply(lambda x:-np.log10(x)),s=4,label="log2FC",color="lime")
plt.scatter(x=colon_df['log2FoldChange'],y=colon_df['padj'].apply(lambda x:-np.log10(x)),s=1,label="Not significant",color="grey")


texts=[]
for i,r in filtered_colon_df.iterrows():
    if r['log2FoldChange'] > 2:
        texts.append(plt.text(x=r['log2FoldChange'],y=-np.log10(r['padj']),s=i,color='Orange'))
    elif r['log2FoldChange'] < -2: 
        texts.append(plt.text(x=r['log2FoldChange'],y=-np.log10(r['padj']),s=i,color='Blue'))

# for i,r in filtered_colon_df.iterrows():
#     texts.append(plt.text(x=r['log2FoldChange'],y=-np.log10(r['padj']),s=i))
# for i,r in padj_colon_df.iterrows():
#     texts.append(plt.text(x=r['log2FoldChange'],y=-np.log10(r['padj']),s=i))
# for i,r in log2fc_colon_df.iterrows():
#     texts.append(plt.text(x=r['log2FoldChange'],y=-np.log10(r['padj']),s=i))
    
    
adjust_text(texts,arrowprops=dict(arrowstyle="-", color='black', lw=0.2))



plt.xlabel("log2FC")
plt.ylabel("log10padj")
plt.axvline(-2,color="grey",linestyle="--", lw=0.3)
plt.axvline(2,color="grey",linestyle="--", lw=0.3)
plt.axhline(2,color="grey",linestyle="--", lw=0.3)
# plt.legend(loc='upper right')
plt.savefig(os.path.abspath(os.path.join(HERE, os.pardir, 'results', 'DeSeq2','meta_species', 'Colon_Digesta_35_DC_CTR_volcano.png')))
```
