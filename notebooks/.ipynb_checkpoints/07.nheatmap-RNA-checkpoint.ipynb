```python
# Function from absolute abundances to clr
def to_clr(data):
    data += 1                                  # add pseudocount
    data = data.div(data.sum(axis=0), axis=1)  # relative abund
    return pd.DataFrame(clr(data.T), columns=data.index, index=data.columns).T #clr
```

```python
digesta_raw_species_clr_T=to_clr(digesta_raw_species.T)
digesta_swab_marked_raw_species_clr_T=to_clr(digesta_swab_marked_raw_species.T)
```

```python
import pandas as pd
from skbio.stats.composition import clr


def variance_threshold_selector(data, threshold=1):
    selector = VarianceThreshold(threshold)
    selector.fit(data)
    return data[data.columns[selector.get_support(indices=True)]]
```
