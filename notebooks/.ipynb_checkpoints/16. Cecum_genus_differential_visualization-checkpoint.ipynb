```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
time_points=['3','14','21','35']
day_points=['Day35', 'Day21', 'Day14', 'Day3']
treatments=['AGP', 'DC']
```

```python
taxonomy_result_path='/home/chengyao/Documents/Phd_project/chicken_microbiome/DEanalysis/results/metagenome/taxonomy/'
functional_result_path='/home/chengyao/Documents/Phd_project/chicken_microbiome/DEanalysis/results/metagenome/functional/'
host_RNA_result_path='/home/chengyao/Documents/Phd_project/chicken_microbiome/DEanalysis/results/RNA_seq/'
```

```python
def up_down(path, where):

    down={}
    up={}

    for time in time_points:
        day='Day'+time
        down[day]={}
        up[day]={}

        for treatment in treatments:
            file_name = path + where + "_" + time + "_" + treatment+'_CTR_shrunk_stat_results.xlsx'
            shrunk_results = pd.read_excel(file_name)
            down_num=len(shrunk_results[shrunk_results['log2FoldChange'] < -1])
            up_num=len(shrunk_results[shrunk_results['log2FoldChange'] > 1])

            down[day][treatment]=down_num
            up[day][treatment]=up_num
            
    down_df = pd.DataFrame(down).T
    up_df = pd.DataFrame(up).T
    
    return down_df, up_df
```

```python
metaG_digesta_species_down_df, metaG_digesta_species_up_df = up_down(taxonomy_result_path, 'Cecum_Digesta')
metaG_mucosa_species_down_df, metaG_mucosa_species_up_df = up_down(taxonomy_result_path, 'Cecum_Mucosa')
```

```python
metaG_digesta_ko_down_df, metaG_digesta_ko_up_df = up_down(functional_result_path, 'Cecum_Digesta')
metaG_mucosa_ko_down_df, metaG_mucosa_ko_up_df = up_down(functional_result_path, 'Cecum_Mucosa')
```

```python
host_mucosa_down_df, host_mucosa_up_df = up_down(host_RNA_result_path, 'Cecum_Mucosa')
```

```python
curchangesDOWN = metaG_digesta_species_down_df.loc[day_points, 'DC']
curchangesUP = metaG_digesta_species_up_df.loc[day_points, 'DC']

fig, axes = plt.subplots(ncols=2, sharey=True)
axes[0].barh(day_points, [-x for x in curchangesDOWN], align='center', color='#0073e6')
axes[0].invert_yaxis()
axes[0].set(yticks=day_points)
axes[0].set_xlim(-30,0)
axes[0].set_ylabel('Time',fontsize=15)
axes[0].set_title('Down',fontsize=15)
axes[0].tick_params(axis='y', which='major', labelsize=15)
axes[1].barh(day_points, curchangesUP, align='center', color='#c44601')
axes[1].set_xlim(0,30)
axes[1].set_title('Up',fontsize=15)
axes[1].invert_yaxis()

plt.subplots_adjust(wspace=0, hspace=0)
plt.tick_params(left = False) 


plt.savefig('../results/DeSeq2/metaG_digesta_species_DC.png')
```
