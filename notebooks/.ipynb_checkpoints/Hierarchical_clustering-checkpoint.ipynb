```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os 
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
from nheatmap import nhm, scripts
from matplotlib.colors import LinearSegmentedColormap

```

```python
# Paths
HERE = os.getcwd()
DATA_DIRECTORY = os.path.abspath(os.path.join(HERE, os.pardir, 'data'))

MICROBIOME_PATH = os.path.join(DATA_DIRECTORY, 'interim', 'metagenome', 'taxonomy', 'new_GTDB_cecum_species_relative_abundance_filterd_low.xlsx')
HOST_TRANSCRIPTOME_PATH = os.path.join(DATA_DIRECTORY, 'interim', 'RNA_seq', 'rna_seq_filtered_low_rld_euclidean.xlsx')
METADATA_PATH = os.path.join(DATA_DIRECTORY, 'interim', 'metadata.xlsx')
```

# 1. Microbiome

```python
metadata = pd.read_excel(METADATA_PATH, header=0, index_col=0)
output_relative_species_excel = pd.read_excel (MICROBIOME_PATH, index_col=0)
```

## 1.1 Bray-curtis distance calculation

```python
microbiome = pd.read_excel(MICROBIOME_PATH, index_col=0)
microbiome = microbiome.loc[:, (output_relative_species_excel!= 0).any(axis=0)]
```

```python
three_basic_treatments_bray_curtis_condensed = pdist(microbiome,metric='braycurtis')
three_basic_treatments_bray_curtis_full = squareform(three_basic_treatments_bray_curtis_condensed)
three_basic_treatments_bray_curtis_full_df = pd.DataFrame(three_basic_treatments_bray_curtis_full, index = output_relative_species_excel.index, columns=output_relative_species_excel.index)
three_basic_treatments_bray_curtis_full_df
```

## 1.2 Hierarchical clustering

```python
three_basic_treatsment_type_cecum_dfrc = metadata[['AGE','TREATMENT','TYPE']]

replace_dict = {"Digesta": "Digesta",
               "Mucosa/Tissue": "Mucosa",
               "CTR": "1. CTR",
               "AGP": "2. AGP",
               "PFA": "3. PFA",
               3: "Day03",
               14: "Day14",
               21: "Day21",
               35: "Day35"}

three_basic_treatsment_type_cecum_dfrc = three_basic_treatsment_type_cecum_dfrc.replace(to_replace=replace_dict)
```

```python

# Define your own colors
colors_treatment = ["lightgrey", "gold", "steelblue"]  # Blue, Green, Yellow, Red
colors_type=["#fdae61","#abd9e9"]

# Create a custom colormap
custom_treatment = LinearSegmentedColormap.from_list("custom_cmap", colors_treatment, N=3)
custom_type = LinearSegmentedColormap.from_list("custom_cmap", colors_type, N=2)

cmaps={'AGE':'Purples', 'TREATMENT':custom_treatment, 'TYPE':custom_type}

cmap = plt.get_cmap('magma')
OrRd_half = LinearSegmentedColormap.from_list(
    'OrRd_half', cmap(np.linspace(0, 1, 256)))
```

```python
g = nhm(data=three_basic_treatments_bray_curtis_full_df, dfc=three_basic_treatsment_type_cecum_dfrc, figsize=(10, 8), linewidths=0, showxticks=False, showyticks=False,cmaps=cmaps, cmapCenter=OrRd_half)
g.hcluster(method='average',optimal_ordering=True)
fig, plots = g.run()
fig.savefig('../figures/microbiome_heatmap.png', bbox_inches='tight')
```

# 2. Host transcriptome

```python
## rlog transformation and euclidean distance was calculated in R
host_transcriptome_euclidean = pd.read_excel(HOST_TRANSCRIPTOME_PATH, index_col=0)
host_transcriptome_euclidean
```

```python
metadata_selected=metadata.loc[host_transcriptome_euclidean.index]
metadata_selected=metadata_selected[['AGE',"TREATMENT"]]
replace_dict = {3: "Day03",
               14: "Day14",
               21: "Day21",
               35: "Day35"}

metadata_selected = metadata_selected.replace(to_replace=replace_dict)
```

```python
g = nhm(data=host_transcriptome_euclidean, dfc=metadata_selected, figsize=(10, 8), linewidths=0, showxticks=False, showyticks=False,cmaps=cmaps, cmapCenter=OrRd_half)
g.hcluster(method='average',optimal_ordering=True)
fig, plots = g.run()
fig.savefig('../figures/host_transcriptome_heatmap.png', bbox_inches='tight')
```
