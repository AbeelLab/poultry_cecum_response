```python
import pandas as pd
import matplotlib
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np
import os



HERE = os. getcwd()

species_RESULT_DIRECTORY = os.path.abspath(os.path.join(HERE, os.pardir, 'DEanalysis', 'results_time_course', 'time_courseLRT', 'digesta_species'))
host_mucosa_RNA_RESULT_DIRECTORY = os.path.abspath(os.path.join(HERE, os.pardir, 'DEanalysis', 'results_time_course', 'time_courseLRT', 'host_mucosa_RNA'))

OUTPUT_PATH = os.path.abspath(os.path.join(HERE, os.pardir, 'DEanalysis', 'results_time_course', 'time_courseLRT'))
```

```python
species_normalized_counts = pd.read_csv(os.path.join(species_RESULT_DIRECTORY, 'cecum_digesta_normalized_counts.txt'), sep='\t', index_col=0)
```

```python
species_normalized_counts
```

```python

host_mucosa_RNA_normalized_counts= pd.read_excel(os.path.join(host_mucosa_RNA_RESULT_DIRECTORY, 'normalized_counts.xlsx'), index_col=0)
host_mucosa_RNA_normalized_counts
```

# AGP

```python
AGP_sig_species = pd.read_excel(os.path.join(species_RESULT_DIRECTORY, 'AGP', 'AGP_sig.xlsx'))
AGP_sig_species
```

```python
sig_species_normalized_counts=species_normalized_counts[species_normalized_counts.index.isin(AGP_sig_species.species.values.tolist())]
```

```python
AGP_sig_species_normalized_counts = sig_species_normalized_counts.filter(regex='C').T
AGP_sig_species_normalized_counts
```

```python
from scipy import stats
AGP_sig_species_normalized_counts_z= stats.zscore(AGP_sig_species_normalized_counts,axis=0)
AGP_sig_species_normalized_counts_z
```

```python
AGP_sig_species_normalized_counts_z.to_excel(os.path.join(species_RESULT_DIRECTORY, 'AGP_cecum_digesta_normalized_counts_zscore.xlsx'), index_label='sample_id')
```

```python
AGP_sig_host_mucosa_RNA = pd.read_excel(os.path.join(host_mucosa_RNA_RESULT_DIRECTORY, 'AGP', 'AGP_sig.xlsx'))
AGP_sig_host_mucosa_RNA
```

```python
sig_host_mucosa_RNA_normalized_counts=host_mucosa_RNA_normalized_counts[host_mucosa_RNA_normalized_counts.index.isin(AGP_sig_host_mucosa_RNA.gene.values.tolist())]
```

```python
AGP_sig_host_mucosa_RNA_normalized_counts = sig_host_mucosa_RNA_normalized_counts.filter(regex='C').T
AGP_sig_host_mucosa_RNA_normalized_counts
```

```python
AGP_sig_host_mucosa_RNA_normalized_counts_z= stats.zscore(AGP_sig_host_mucosa_RNA_normalized_counts,axis=0)
AGP_sig_host_mucosa_RNA_normalized_counts_z
```

```python

AGP_sig_host_mucosa_RNA_normalized_counts_z.to_excel(os.path.join(host_mucosa_RNA_RESULT_DIRECTORY, 'AGP_cecum_host_mucosa_RNA_normalized_counts_zscore.xlsx'), index_label='sample_id')
```

# DC

```python
DC_sig_species = pd.read_excel(os.path.join(species_RESULT_DIRECTORY, 'DC', 'DC_sig.xlsx'))
sig_species_normalized_counts=species_normalized_counts[species_normalized_counts.index.isin(DC_sig_species.species.values.tolist())]
DC_sig_species_normalized_counts = sig_species_normalized_counts.filter(regex='B').T
DC_sig_species_normalized_counts
```
