```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

ko_id_name_table_path = "/home/chengyao/Documents/Phd_project/chicken_microbiome/data/raw/metagenomic_data/KO.txt"


taxonomy_result_path='/home/chengyao/Documents/Phd_project/chicken_microbiome/DEanalysis/results_differential/metagenome/taxonomy/'
functional_result_path='/home/chengyao/Documents/Phd_project/chicken_microbiome/DEanalysis/results_differential/metagenome/functional/'
host_RNA_result_path='/home/chengyao/Documents/Phd_project/chicken_microbiome/DEanalysis/results_differential/RNA_seq/'
```

```python
time_points=['3','14','21','35']
day_points=['Day35', 'Day21', 'Day14', 'Day3']
treatments=['AGP', 'DC']
```

```python
ko_id_name_table=pd.read_csv(ko_id_name_table_path, sep=r'\;|\t|\[EC', names=['id', 'name', 'description','EC'])
ko_id_name_table['combined'] = ko_id_name_table['id']+ ': '+ ko_id_name_table['name'] +"; "+ ko_id_name_table['description']
ko_id_name_dict=dict(zip(ko_id_name_table['id'], ko_id_name_table['combined']))
ko_id_name_dict
# for key in ko_id_name_dict.keys():
#     if not ko_id_name_dict[key] is None:
#         ko_id_name_dict[key] = ko_id_name_dict[key].strip()
# ko_id_name_dict
```

```python
def differential_summary_df(path, where, id_flag):
    results = []

    for time in time_points:
        day='Day'+time

        for treatment in treatments:
            if id_flag:
                file_name = path + where + "_" + time + "_" + treatment+'_CTR_shrunk_stat_results_id.xlsx'
            else:
                file_name = path + where + "_" + time + "_" + treatment+'_CTR_shrunk_stat_results.xlsx'
            shrunk_results = pd.read_excel(file_name, index_col=0)
            for i in range(len(shrunk_results)):
                species=shrunk_results.index[i]
                log2FoldChange=shrunk_results.loc[species, 'log2FoldChange']
                results.append([day, treatment, species, log2FoldChange])
    return results
            
        
```

```python
metaG_digesta_species = differential_summary_df(taxonomy_result_path, 'Cecum_Digesta', False)
metaG_mucosa_species = differential_summary_df(taxonomy_result_path, 'Cecum_Mucosa', False)
metaG_digesta_ko = differential_summary_df(functional_result_path, 'Cecum_Digesta', True)
metaG_mucosa_ko = differential_summary_df(functional_result_path, 'Cecum_Mucosa', True)
host_mucosa_expression = differential_summary_df(host_RNA_result_path, 'Cecum_Mucosa', False)
```

```python
metaG_digesta_species_df = pd.DataFrame(metaG_digesta_species, columns=['Age', 'Treatment', 'Species', 'Log2FoldChange'])
metaG_mucosa_species_df = pd.DataFrame(metaG_mucosa_species, columns=['Age', 'Treatment', 'Species', 'Log2FoldChange'])
metaG_digesta_ko_df = pd.DataFrame(metaG_digesta_ko, columns=['Age', 'Treatment', 'KO group', 'Log2FoldChange'])
metaG_mucosa_ko_df = pd.DataFrame(metaG_mucosa_ko, columns=['Age', 'Treatment', 'KO group', 'Log2FoldChange'])
host_mucosa_expression_df = pd.DataFrame(host_mucosa_expression, columns=['Age', 'Treatment', 'Gene', 'Log2FoldChange'])
```
