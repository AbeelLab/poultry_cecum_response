```python
import os
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Paths
HERE = os.getcwd()
DATA_DIRECTORY = os.path.abspath(os.path.join(HERE, os.pardir, 'data'))

OUTPUT_RELATIVE_SPECIES_PATH = os.path.join(DATA_DIRECTORY, 'interim', 'metagenome', 'taxonomy', 'new_GTDB_cecum_species_relative_abundance_filterd_low.xlsx')
METADATA_PATH = os.path.join(DATA_DIRECTORY, 'interim', 'metadata.xlsx')
```

```python
# Load the relative abundance microbiome data and metadata
output_relative_species = pd.read_excel(OUTPUT_RELATIVE_SPECIES_PATH, index_col=0)
metadata=pd.read_excel(METADATA_PATH, index_col=0)
metadata
```

```python
def topn_plot(day, sample_type, topn=10, df=output_relative_species):
    """
    Function to plot the topn species based on their median relative abundance 
    for a given day and sample type.
    """

    
    filtered_metadata = metadata[
        (metadata['AGE'] == day) & 
        (metadata['TYPE'] == sample_type)
    ]

    species_day_sample_type = df.loc[filtered_metadata.index]
    species_sorted = species_day_sample_type.median().sort_values(ascending=True).index
    species_sorted_topn = species_sorted[-topn:]

    species_day_sample_type_topn = species_day_sample_type[species_sorted_topn]
    
    # Plot
    ax = species_day_sample_type_topn.plot.box(
        vert=False, figsize=(7, 5), color='black', showfliers=False
    )
    ax.set_xlabel('Relative abundance', fontsize=13)
    ax.set_ylabel('Species', fontsize=13)
    ax.set_xlim(0, 0.42)
    plt.yticks(fontsize=13)
    ax.axvline(x=0, color="r", linestyle="--")
    
    plt.title(f'Top {topn} species for {sample_type} on Day {day}')
    plt.savefig(f'../figures/{day}_{sample_type}_{topn}_cecum_species_relative_abundance.png', bbox_inches='tight')
    plt.show()
```

```python
# Plot for different days and sample types
days = [3, 14, 21, 35]
sample_types = ['Digesta', 'Mucosa']

for day in days:
    for sample_type in sample_types:
        topn_plot(day, sample_type, topn=10)
```
