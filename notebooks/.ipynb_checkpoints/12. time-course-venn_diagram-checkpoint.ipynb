# 1. 14 day

```python
def gene_list(gut_section, day, treatments):
    gene_list_dict={}
    for treatment in treatments:
        treatment_table = pd.read_excel(os.path.join(OUTPUT_PATH, gut_section+"_Mucosa_"+day+"_"+treatment+"_CTR_shrunk_stat_results.xlsx"),index_col=0)
        filtered_treatment_table = treatment_table[treatment_table['padj'] < 0.01]
        gene_list_dict[treatment]=set(filtered_treatment_table.index.tolist())
    return gene_list_dict

        
```

```python
ileum_14=gene_list('Ileum', '14', ['AGP','DC','HBT'])
venny4py(sets=ileum_14,out=OUTPUT_PATH)

```

```python
cecum_14 = gene_list('Cecum', '14', ['AGP','DC','HBT'])
venny4py(sets=cecum_14,out=OUTPUT_PATH)
```

```python
colon_14=gene_list('Colon', '14', ['AGP','DC','HBT'])
venny4py(sets=colon_14,out=OUTPUT_PATH)
```

# 4. All times

```python
def merge_dictionaries(dict1, dict2, dict3):
    merged_dict = {}
    for key in dict1.keys():
        merged_dict[key] = dict1[key].union(dict2[key], dict3[key])
    return merged_dict
```

```python
ileum_all_times=merge_dictionaries(ileum_14,ileum_21,ileum_35)
venny4py(sets=ileum_all_times,out=OUTPUT_PATH)
```
