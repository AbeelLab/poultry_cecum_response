```python
import pandas as pd
import matplotlib
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np
import os



HERE = os. getcwd()

species_RESULT_DIRECTORY = os.path.abspath(os.path.join(HERE, os.pardir, 'DEanalysis', 'results_time_course', 'time_courseLRT', 'digesta_species'))
host_mucosa_RNA_RESULT_DIRECTORY = os.path.abspath(os.path.join(HERE, os.pardir, 'DEanalysis', 'results_time_course', 'time_courseLRT', 'host_mucosa_RNA'))

OUTPUT_PATH = os.path.abspath(os.path.join(HERE, os.pardir, 'DEanalysis', 'results_time_course', 'time_courseLRT'))
```

```python
species_normalized_counts = pd.read_csv(os.path.join(species_RESULT_DIRECTORY, 'cecum_digesta_normalized_counts.txt'), sep='\t', index_col=0)
```

```python
species_normalized_counts
```

```python

host_mucosa_RNA_normalized_counts= pd.read_excel(os.path.join(host_mucosa_RNA_RESULT_DIRECTORY, 'normalized_counts.xlsx'), index_col=0)
host_mucosa_RNA_normalized_counts
```

# AGP

```python
AGP_sig_species = pd.read_excel(os.path.join(species_RESULT_DIRECTORY, 'AGP', 'AGP_sig.xlsx'))
AGP_sig_species
```

```python
sig_species_normalized_counts=species_normalized_counts[species_normalized_counts.index.isin(AGP_sig_species.species.values.tolist())]
```

```python
days = {'D3':'Day3', 'D14': 'Day14', 'D21': 'Day21', 'D35':'Day35'}
AGP_sig_species_normalized_counts = sig_species_normalized_counts.filter(regex='B')

listofdf=[]
for key in days.keys():
    listofdf.append(pd.DataFrame(AGP_sig_species_normalized_counts.filter(regex=key).sum(axis=1), columns=[[days[key]]]))

AGP_species = pd.concat(listofdf, axis=1)
AGP_species.shape
```

```python
AGP_sig_host_mucosa_RNA = pd.read_excel(os.path.join(host_mucosa_RNA_RESULT_DIRECTORY, 'AGP', 'AGP_sig.xlsx'))
```

```python
sig_host_mucosa_RNA_normalized_counts=host_mucosa_RNA_normalized_counts[host_mucosa_RNA_normalized_counts.index.isin(AGP_sig_host_mucosa_RNA.species.values.tolist())]
```

```python
sig_host_mucosa_RNA_normalized_counts
```

```python
days = {'M3':'Day3', 'M14': 'Day14', 'M21': 'Day21', 'M35':'Day35'}
AGP_sig_host_mucosa_RNA_normalized_counts = sig_host_mucosa_RNA_normalized_counts.filter(regex='B')
AGP_sig_host_mucosa_RNA_normalized_counts
listofdf=[]
for key in days.keys():
    listofdf.append(pd.DataFrame(AGP_sig_host_mucosa_RNA_normalized_counts.filter(regex=key).sum(axis=1), columns=[[days[key]]]))

AGP_host_mucosa_RNA = pd.concat(listofdf, axis=1)
AGP_host_mucosa_RNA
```

```python
AGP_res = scipy.stats.spearmanr(a=AGP_species, b=AGP_host_mucosa_RNA, axis=1, nan_policy='propagate', alternative='two-sided')
```

```python
AGP_species_num = AGP_species.shape[0]
statistics = pd.DataFrame(AGP_res.statistic).iloc[AGP_species_num:,:AGP_species_num]
statistics.index = AGP_host_mucosa_RNA.index.tolist()
statistics.columns = AGP_species.index.tolist()
statistics
```

```python
pvalue = pd.DataFrame(AGP_res.pvalue).iloc[AGP_species_num:,:AGP_species_num]
pvalue.index = AGP_host_mucosa_RNA.index.tolist()
pvalue.columns = AGP_species.index.tolist()
pvalue=pvalue.T
pvalue
```

```python
shape = pvalue.values.shape
significant_matrix = multipletests(pvalue.values.flatten())[0].reshape(shape)
significant_matrix.shape
```

```python
spearman_data = statistics.T
spearman_data
```

```python
spearman_data.shape
```

```python
g = sns.clustermap(spearman_data, cmap="bwr", vmin=-1, vmax=1,yticklabels=True)
for tick in g.ax_heatmap.get_xticklabels():
    tick.set_rotation(30)


plt.figure(figsize=(20,25))



for i, ix in enumerate(g.dendrogram_row.reordered_ind):
    for j, jx in enumerate(g.dendrogram_col.reordered_ind):
        text = g.ax_heatmap.text(
            j + 0.5,
            i + 0.85,
            "*" if significant_matrix[ix, jx] else "",
            ha="center",
            va="center",
            color="black",
        )
        text.set_fontsize(15)
        
        
g.savefig('../results/AGP_hostRNA_digestaSpecies_hclustering.png', bbox_inches='tight')
```

# PFA

```python
DC_sig_species = pd.read_excel(os.path.join(species_RESULT_DIRECTORY, 'DC', 'DC_sig.xlsx'))
DC_sig_species
```

```python
sig_species_normalized_counts=species_normalized_counts[species_normalized_counts.index.isin(DC_sig_species.species.values.tolist())]
```

```python
days = {'D3':'Day3', 'D14': 'Day14', 'D21': 'Day21', 'D35':'Day35'}
DC_sig_species_normalized_counts = sig_species_normalized_counts.filter(regex='B')

listofdf=[]
for key in days.keys():
    listofdf.append(pd.DataFrame(DC_sig_species_normalized_counts.filter(regex=key).sum(axis=1), columns=[[days[key]]]))

DC_species = pd.concat(listofdf, axis=1)
DC_species.shape
```

```python
DC_sig_host_mucosa_RNA = pd.read_excel(os.path.join(host_mucosa_RNA_RESULT_DIRECTORY, 'DC', 'DC_sig.xlsx'))
```

```python
sig_host_mucosa_RNA_normalized_counts=host_mucosa_RNA_normalized_counts[host_mucosa_RNA_normalized_counts.index.isin(DC_sig_host_mucosa_RNA.species.values.tolist())]
```

```python
sig_host_mucosa_RNA_normalized_counts
```

```python
days = {'M3':'Day3', 'M14': 'Day14', 'M21': 'Day21', 'M35':'Day35'}
DC_sig_host_mucosa_RNA_normalized_counts = sig_host_mucosa_RNA_normalized_counts.filter(regex='B')
DC_sig_host_mucosa_RNA_normalized_counts
listofdf=[]
for key in days.keys():
    listofdf.append(pd.DataFrame(DC_sig_host_mucosa_RNA_normalized_counts.filter(regex=key).sum(axis=1), columns=[[days[key]]]))

DC_host_mucosa_RNA = pd.concat(listofdf, axis=1)
DC_host_mucosa_RNA
```

```python
DC_res = scipy.stats.spearmanr(a=DC_species, b=DC_host_mucosa_RNA, axis=1, nan_policy='propagate', alternative='two-sided')
```

```python
DC_species_num = DC_species.shape[0]
statistics = pd.DataFrame(DC_res.statistic).iloc[DC_species_num:,:DC_species_num]
statistics.index = DC_host_mucosa_RNA.index.tolist()
statistics.columns = DC_species.index.tolist()
statistics
```

```python
pvalue = pd.DataFrame(DC_res.pvalue).iloc[DC_species_num:,:DC_species_num]
pvalue.index = DC_host_mucosa_RNA.index.tolist()
pvalue.columns = DC_species.index.tolist()
pvalue = pvalue.T

```

```python
shape = pvalue.values.shape
significant_matrix = multipletests(pvalue.values.flatten())[0].reshape(shape)
significant_matrix.shape
```

```python
spearman_data = statistics.T
spearman_data
```

```python
nrows, ncols = np.shape(spearman_data)
pc = sklearn.decomposition.PCA().fit(spearman_data)
dfr = pd.DataFrame(pc.transform(spearman_data)[:, 0], index=spearman_data.index,columns=['PC1'])
dfr
```

```python
g = sns.clustermap(spearman_data, cmap="bwr", vmin=-1, vmax=1,yticklabels=True, tree_kws=dict(linewidths=0.8))
for tick in g.ax_heatmap.get_xticklabels():
    tick.set_rotation(30)


plt.figure(figsize=(20,25))



for i, ix in enumerate(g.dendrogram_row.reordered_ind):
    for j, jx in enumerate(g.dendrogram_col.reordered_ind):
        text = g.ax_heatmap.text(
            j + 0.5,
            i + 0.85,
            "*" if significant_matrix[ix, jx] else "",
            ha="center",
            va="center",
            color="black",
        )
        text.set_fontsize(15)
        
g.savefig('../results/DC_hostRNA_digestaSpecies_hclustering.png', bbox_inches='tight')
```
